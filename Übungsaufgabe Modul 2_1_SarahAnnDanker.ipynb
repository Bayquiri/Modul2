```python
import json
```

```python
import requests
```

```python
url = "https://www.bibsonomy.org/json/search/finanzen?items=1000"
```

```python
result = requests.get(url)
```

```python
data = result.json()
```

```python
type(data)
```

```python
print(json.dumps(data, indent=2))
```

```python
print(data.keys())
```

```python
items = data['items']
print(type(items))
print(len(items))
```

```python
publications = [item for item in items if item['type'] == 'Publication']
print(len(publications))
```

```python
pubs_with_abstracts = [item for item in publications if 'abstract' in item]
print(len(pubs_with_abstracts))
```

```python
from langdetect import detect
```

```python
german_pubs = [item for item in publications if detect(item['label']) == 'de']
len(german_pubs)
```

```python
print(json.dumps(german_pubs, indent=4))
```

#  In Lingo überführen
- Pro Dokument gibt es eine eindeutige ID, die in eckigen Klammern geschrieben wird, numerisch ist und mit einem Punkt endet.
- Darunter folgt Text (Titel, Abstract)
- Datei speichern unter bibsonomy_to_lingo.txt

```python
type(german_pubs)
```

```python
bibsonomy_list = german_pubs
for index, item in enumerate(bibsonomy_list):
        print('['+str(index)+'.]', "Titel:", item["label"], "Autor:", item["user"])
        
```

```python
bibsonomy_list = german_pubs
for index, item in enumerate(bibsonomy_list):
        print('['+str(index)+'.]')
        print("Titel:", item["label"])
        print("Abstract:", item["user"])
```

```python
bibsonomy_list = german_pubs
for index, item in enumerate(bibsonomy_list):
        print('['+str(index)+'.]')
        print('Titel:', item['label'])
        if 'abstract' in bibsonomy_list:
            print('Abstract:', item['abstract'])
        else:
            print('Kein Abstract vorhanden')
        
```
